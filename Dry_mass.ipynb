```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab dane/Lipidy/Sucha masa - Lipidy.csv', decimal=',')
strains = {1:'YAHO', 2:'YAKE', 3:'YALI',4:'YAPO',5:'YADI', 6:'YAPH', 7:'YADE', 8:'YABU', 9:'YAGA', 10:'OLHI', 11:'YAOS', 12:'YAYA', 13:'YABR', 14:'YAAL', 15:'YAPA'}
data['Strain'] = data['Strain'].map(strains)
data=data.rename(columns = {'Unnamed: 1':'Carbon'})
data['d-w']=(data['m_weight']-data['m_pap'])*100
data['Carbon']=data['Carbon'].map({'A':'Acetate','P':'Propionate','B':'Butyrate'})
data
```

```python
g = sns.catplot(kind='bar', data=data, x='Strain', y='d-w', col='Carbon', palette='tab10')
g.set_xticklabels(rotation=90, size =12)
plt.yticks(size=2)
g.set_xlabels(size=14)
g.set_ylabels(size=14)
g.set_titles('{col_name}')
g.set_axis_labels(x_var='Strain', y_var='DCW g/L')
```
