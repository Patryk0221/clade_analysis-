```python
import pandas as pd
import numpy as np
```

```python
parametry_wzrost = pd.read_csv('/content/drive/MyDrive/Colab dane/Biotek pH/parametry_dopasowane2.csv')
parametry_wzrost.drop(columns = 'Unnamed: 0', inplace = True)
parametry_wzrost.head()
```

```python
parametry_wzrost.set_index('Strain')
cols = parametry_wzrost.drop(columns = 'Strain').columns
parametry_biotek = parametry_wzrost.groupby('Strain')[cols].mean()
parametry_biotek.reset_index(inplace = True)
parametry_biotek
```

```python
kinetyka_parametry = pd.read_csv('/content/drive/MyDrive/Colab dane/Kinetyka/Kinetyka_parametry.csv')
kinetyka_parametry
```

```python
parametry = pd.merge(parametry_biotek, kinetyka_parametry, left_on ='Strain', right_on='STRAIN' )
label = list(parametry['Strain'])
parametry = parametry.drop(columns='STRAIN')
parametry = parametry.iloc[:,1:]
parametry.columns
```

```python
def hierarhical_clustering(data, method, num_clusters, labels, dendrogram_title):
  from scipy.cluster.hierarchy import linkage, fcluster, dendrogram, cophenet
  from sklearn.preprocessing import StandardScaler
  from sklearn.metrics import silhouette_score
  scaler = StandardScaler()
  scaler.fit(data)
  samples = scaler.transform(data)
  linkagemergings = linkage(samples, method=method)

  clusters_num = num_clusters

  clusters = fcluster(linkagemergings, clusters_num, criterion='maxclust')

  silhouette_avg = silhouette_score(samples, clusters)

  print(f"Średni współczynnik Silhouette: {silhouette_avg}")
  from scipy.spatial.distance import pdist
  c, coph_dists = cophenet(linkagemergings, pdist(samples))
  c
  print(f"współczynnik cophenet: {c}")
  import matplotlib.pyplot as plt
  dendrogram(linkagemergings,labels=labels ,leaf_rotation=90, leaf_font_size=6,
            color_threshold=9)
  plt.title(dendrogram_title)
  plt.show()
```

```python
hierarhical_clustering(data = parametry, method = 'average', num_clusters = 3, labels = label, dendrogram_title = 'VFAs utylisation' )
```

```python
def hierarhical_clustering_normal(data, method, num_clusters, labels, dendrogram_title, treshold = 6):
  from scipy.cluster.hierarchy import linkage, fcluster, dendrogram, cophenet
  from sklearn.metrics import silhouette_score

  from sklearn.preprocessing import MinMaxScaler

  scaler = MinMaxScaler()

  samples = scaler.fit_transform(data)
  linkagemergings = linkage(samples, method=method)

  clusters_num = num_clusters

  clusters = fcluster(linkagemergings, clusters_num, criterion='maxclust')

  silhouette_avg = silhouette_score(samples, clusters)

  print(f"Średni współczynnik Silhouette: {silhouette_avg}")
  from scipy.spatial.distance import pdist
  c, coph_dists = cophenet(linkagemergings, pdist(samples))
  c
  print(f"współczynnik cophenet: {c}")
  import matplotlib.pyplot as plt
  dendrogram(linkagemergings,labels=labels ,leaf_rotation=90, leaf_font_size=6,
            color_threshold=treshold)
  plt.title(dendrogram_title)
  plt.show()
  pairs = pd.DataFrame({'Strain': labels, 'cluster': clusters})
  return pairs
```

```python
hierarhical_clustering_normal(data = parametry, method = 'average', num_clusters = 4, labels = label, dendrogram_title = 'VFAs all parameters', treshold = 3 )
```
