```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab dane/Piroliza/Lipidy - piroliza.csv')
df.head()
```

```python
plt.figure(figsize=(8, 6))

g = sns.catplot(
    data=df, x='Strain', y='TOTAL %', hue='Medium',
    kind='bar', palette = 'muted'
)
g.set_axis_labels('Strain', '% of total lipids')
g.set_xticklabels(rotation=90)
plt.rc('axes', titlesize=16)
plt.rc('axes', labelsize=14)
plt.rc('xtick', labelsize=12)
plt.rc('ytick', labelsize=12)
plt.rc('legend', fontsize=12)
```

```python
df_types = df.drop(columns ='TOTAL %')
df_melted = pd.melt(df_types, id_vars = df_types.columns[:2], var_name='FA type')
df_melted.head()
```

```python
plt.figure(figsize=(8, 6))

g = sns.catplot(
    data=df_melted, x='Strain', y='value', hue='FA type', col = 'Medium',
    kind='bar', palette = 'muted'
)
g.set_axis_labels('Strain', '% of FA type')
g.set_xticklabels(rotation=90)
plt.rc('axes', titlesize=16)
plt.rc('axes', labelsize=14)
plt.rc('xtick', labelsize=12)
plt.rc('ytick', labelsize=12)
plt.rc('legend', fontsize=12)
```
